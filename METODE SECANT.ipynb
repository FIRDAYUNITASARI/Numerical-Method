```python
import numpy as np
import sympy as sp
from prettytable import PrettyTable
import matplotlib.pyplot as plt

## fungsi persamaan
def f(x):
    return (4*(x**3))-(15*(x**2))+(17*x)-6

# Input tebakan awal, toleransi, dan jumlah iterasi maksimum
x0 = float(input("Masukkan tebakan awal x0: "))
x1 = float(input("Masukkan tebakan awal x1: "))
eps = float(input("Masukkan galat yang diinginkan: "))
N = int(input("Masukkan jumlah iterasi maksimum: "))

# masukkan list
xm_list = [x0, x1]
f_list = [f(x0), f(x1)]

for i in range(2, N):
    if abs((xm_list[i-1]-xm_list[i-2])/(xm_list[i-1])) < eps:
        break

    xm_list.append(xm_list[i-1] - ((f(xm_list[i-1]) * (xm_list[i-1]-xm_list[i-2])) / (f(xm_list[i-1]) - f(xm_list[i-2]))))
    f_list.append(f(xm_list[i]))

#Tabel cantik
columns = ["iterasi", "xr", "f(xr)"]

Table = PrettyTable()

# tambahkan kolom
Table.add_column(columns[0], range(1, i+1))
Table.add_column(columns[1], [round(num, 5) for num in xm_list][:i])
Table.add_column(columns[2], [round(num, 5) for num in f_list][:i])

print(Table)

# Plot nilai xr dan iterasinya
plt.plot(range(1, i+1), xm_list[:i], 'go-')
plt.xlabel('Iterasi')
plt.ylabel('Nilai xr')
plt.title('Plot Nilai xr dan Iterasi')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()

```
