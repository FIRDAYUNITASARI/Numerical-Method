```python
import numpy as np
import matplotlib.pyplot as plt
import random

def f(x, y):
    return y * x - x**2

def trapezoidal_rule(f, a, b, n, y):
    h = (b - a) / n
    integral = (f(a, y) + f(b, y)) / 2

    for i in range(1, n):
        x = a + i * h
        integral += f(x, y)

    integral *= h
    return integral

# Generate random value for y between 5 and 40
y = random.uniform(5, 40)
print('nilai y : ', y)

# Define the limits of integration
a = 0
b = 1

# Number of segments
n = 2
print('banyaknya segmen :', n)

# Calculate the approximation using trapezoidal rule
approximation = trapezoidal_rule(f, a, b, n, y)

print("Pendekatan integral:", approximation)

# Plotting
x_vals = np.linspace(a, b, 100)
y_vals = [f(x, y) for x in x_vals]

plt.plot(x_vals, y_vals, 'b-', label='f(x, y)')
plt.fill_between(x_vals, y_vals, color='blue', alpha=0.2, label='Luasan Daerah')

plt.xlabel('Nilai x')
plt.ylabel('Nilai y')
plt.title('Grafik Luasan Daerah')
plt.legend()
plt.show()
```
