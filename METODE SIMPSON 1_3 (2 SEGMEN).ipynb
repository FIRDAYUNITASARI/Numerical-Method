```python
import random
import matplotlib.pyplot as plt
import numpy as np

def f(x, z):
    return 4 * x - x**z

def simpson_rule(f, a, b, n, z):
    h = (b - a) / n
    print('nilai h : ', h)
    integral = f(a, z) + f(b, z)

    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            integral += 2 * f(x, z)
        else:
            integral += 4 * f(x, z)

    integral *= h / 3
    return integral

# Generate random power between 2 and 8
z = random.randint(2, 8)
print('nilai z adalah : ', z)

# Define the limits of integration
a = 0
b = 1

# Number of segments
n = int(input('masukkan banyaknya segmen:'))

# Calculate the approximation using Simpson's rule
approximation = simpson_rule(f, a, b, n, z)
print("Pendekatan integral:", approximation)

# Generate x values for the area plot
x = np.linspace(a, b, 100)
y = f(x, z)

# Plotting the area under the curve
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f(x, z)')
plt.fill_between(x, y, alpha=0.5, label='area')
plt.xlabel('x')
plt.ylabel('f(x, z)')
plt.title('Grafik f(x, z) dan Luasan Daerah')
plt.legend()
plt.show()
```
