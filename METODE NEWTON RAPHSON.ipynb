```python
import math
import numpy as np
import matplotlib.pyplot as plt
from prettytable import PrettyTable

def f(x):
    return math.exp(-x) - x

def f_derivative(x):
    return -math.exp(-x) - 1

def newton_raphson(x0, epsilon):
    x = x0
    iterasi = 0

    table = PrettyTable()
    table.field_names = ["iterasi", "xr", "|xr+1 - xr|"]
    xr_values = []
    iteration_values = []

    while True:
        fx = f(x)
        fx_derivative = f_derivative(x)

        x_next = x - fx / fx_derivative
        iterasi += 1

        error = abs(x_next - x)
        table.add_row([iterasi, x, error])

        if error < epsilon or iterasi >= max_iterasi:
            break

        x = x_next
        xr_values.append(x)
        iteration_values.append(iterasi)

    print(table)
    return xr_values, iteration_values

awal = float(input("Masukkan rentang awal x0: "))
akhir = float(input("Masukkan rentang akhir x0: "))
x0 = np.random.uniform(awal, akhir)
print('sehingga nilai x0 : ', x0)
epsilon = float(input("Masukkan nilai toleransi error: "))
max_iterasi = 100

xr_values, iteration_values = newton_raphson(x0, epsilon)
result = xr_values[-1]
print("Nilai x yang memenuhi f(x) = 0: ", result)

# Plot line plot dan scatter plot
plt.figure(figsize=(8, 6))
plt.plot(iteration_values, xr_values, 'b-', label='Metode Newton-Raphson')
plt.scatter(iteration_values, xr_values, color='red', label='Iterasi')
plt.xlabel('Iterasi')
plt.ylabel('xr')
plt.title('Grafik Metode Newton-Raphson')
plt.legend()
plt.grid(True)
plt.show()
```
