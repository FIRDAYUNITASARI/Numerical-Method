```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
from prettytable import PrettyTable

# Input nilai x dan f(x)
n = int(input("Masukkan jumlah data: "))

x = []
fx = []

for i in range(n):
    xi = float(input(f"Masukkan nilai x_{i+1}: "))
    fxi = float(input(f"Masukkan nilai f(x_{i+1}): "))
    x.append(xi)
    fx.append(fxi)

# Interpolasi spline
spline = CubicSpline(x, fx)

# Tabel perhitungan
columns = ["x", "f(x)", "f'(x)", "f''(x)"]

Table = PrettyTable()

# tambahkan kolom
Table.add_column(columns[0], x)
Table.add_column(columns[1], fx)
Table.add_column(columns[2], spline(x, 1))
Table.add_column(columns[3], spline(x, 2))

print(Table)

# Input nilai x yang akan diinterpolasi
x_interp = float(input('Masukkan nilai x yang akan diinterpolasi: '))

# Melakukan interpolasi pada nilai x yang diberikan
y_interp = spline(x_interp)

# Menampilkan hasil interpolasi
print("Hasil Interpolasi: ", y_interp)

# Plot hasil interpolasi spline
plt.figure(figsize=(8, 6))
plt.scatter(x_interp, y_interp, color='green', label='Hasil Interpolasi')
plt.scatter(x, fx, color='red', label='Data Asli')
plt.plot(x, spline(x), linestyle='-', color='blue', label='Interpolasi Spline')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Interpolasi Spline')
plt.legend()
plt.grid(True)

plt.show()

```
