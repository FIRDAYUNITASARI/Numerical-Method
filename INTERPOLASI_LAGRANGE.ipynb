```python
import matplotlib.pyplot as plt
import numpy as np
def lagrange_interpolation(points, x):
    n = len(points)

    # Membuat fungsi basis Lagrange
    def basis_lagrange(k):
        def basis_func(x):
            result = 1.0
            for i in range(n):
                if i != k:
                    result *= (x - points[i][0]) / (points[k][0] - points[i][0])
            return result
        return basis_func

    # Menghitung nilai interpolasi
    interpolation = 0.0
    for k in range(n):
        interpolation += points[k][1] * basis_lagrange(k)(x)

    return interpolation


# Input data titik-titik interpolasi dari pengguna
n = int(input("Masukkan jumlah titik interpolasi: "))
points = []
for i in range(n):
    x = float(input(f"Masukkan x{i+1}: "))
    y = float(input(f"Masukkan y{i+1}: "))
    points.append((x, y))

# Input nilai x yang akan diinterpolasi
x = float(input("Masukkan nilai x yang akan diinterpolasi: "))

# Menghitung dan menampilkan hasil interpolasi Lagrange
interpolation = lagrange_interpolation(points, x)
print(f"Hasil interpolasi Lagrange pada x = {x}: {interpolation:.4f}")

# Menyiapkan data untuk plot
x_vals = np.linspace(min([p[0] for p in points]), max([p[0] for p in points]), 100)
y_vals = [lagrange_interpolation(points, val) for val in x_vals]

# Plot data asli
plt.scatter([p[0] for p in points], [p[1] for p in points], color='red', label='Data Asli')

# Plot hasil interpolasi
plt.plot(x_vals, y_vals, color='blue', label='Interpolasi Lagrange')

# Plot scatter plot untuk hasil interpolasi pada nilai x yang dicari
plt.scatter(x, interpolation, color='green', label='Interpolasi pada x')

plt.xlabel('Nilai x')
plt.ylabel('Nilai y')
plt.title('Interpolasi Lagrange')
plt.legend()
plt.show()

```
