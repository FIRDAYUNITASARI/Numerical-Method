```python
def deret_taylor(f, x0, y0, h, x):

    n = int((x - x0) / h)  # Jumlah iterasi

    for i in range(n):
        k1 = h * f(x0, y0)
        k2 = h * f(x0 + (h / 2), y0 + (k1 / 2))
        k3 = h * f(x0 + h, y0 - k1 + 2 * k2)
        y0 = y0 + (1 / 6) * (k1 + 4 * k2 + k3)
        x0 = x0 + h

    return y0

def f(x, y):
    return 1 + y**2

# Input nilai awal
x0 = float(input('masukkan nilai x(0) : '))
y0 = float(input('masukkan nilai y(0) : '))

# Input ukuran langkah
h = float(input('masukkan nilai h : '))

# Input nilai x yang dicari y(x)
x = float(input('masukkan nilai x yang di cari : '))

# Memanggil fungsi runge_kutta_method untuk menyelesaikan persamaan diferensial
hasil = deret_taylor(f, x0, y0, h, x)

print(f"Nilai y({x}) = {hasil}")
```
