```python
import random

def gauss_elimination(matriks, vektor):
    n = len(matriks)

    # Menampilkan matriks awal
    print("Matriks Awal:")
    print_matriks(matriks, vektor)
    print()

    # Menjalankan eliminasi Gauss
    for i in range(n):
        if matriks[i][i] == 0.0:
            raise Exception("Pembagian oleh nol terdeteksi.")

        pivot = matriks[i][i]

        for j in range(i, n):
            matriks[i][j] /= pivot

        vektor[i] /= pivot

        for j in range(i+1, n):
            rasio = matriks[j][i]

            for k in range(i, n):
                matriks[j][k] -= rasio * matriks[i][k]

            vektor[j] -= rasio * vektor[i]

        # Menampilkan matriks setelah setiap langkah eliminasi
        print(f"Langkah {i+1}:")
        print_matriks(matriks, vektor)
        print()

    # Solusi mundur (back substitution)
    solusi = [0] * n
    solusi[n-1] = vektor[n-1]

    for i in range(n-2, -1, -1):
        solusi[i] = vektor[i]

        for j in range(i+1, n):
            solusi[i] -= matriks[i][j] * solusi[j]

    # Menampilkan solusi
    print("Solusi:")
    print(solusi)

def print_matriks(matriks, vektor):
    n = len(matriks)
    for i in range(n):
        baris = [f"{x:.2f}" for x in matriks[i]]
        print(" ".join(baris) + " | " + f"{vektor[i]:.2f}")

# Input matriks dari pengguna
n = int(input("Masukkan ukuran matriks (n): "))
matriks = []
print("Masukkan elemen-elemen matriks:")
for i in range(n):
    baris = list(map(float, input().split()))
    matriks.append(baris)

# Input vektor dari pengguna
print("Masukkan elemen-elemen vektor:")
vektor = []

# Input vektor baris ke-1 (acak)
rentang_angka = input("Masukkan rentang angka untuk vektor baris ke-1 (pisahkan dengan spasi): ").split()
vektor.extend([random.uniform(float(rentang_angka[0]), float(rentang_angka[1]))])

# Input vektor baris ke-2 dan ke-3
vektor.extend(list(map(float, input("masukkan elemen vektor baris ke-2 dan ke-3 : ").split())))

gauss_elimination(matriks, vektor)
```
