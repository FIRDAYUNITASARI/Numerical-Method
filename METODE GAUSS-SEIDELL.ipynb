```python
def gauss_seidel(matriks, vektor, nilai_awal, max_iterasi):
    n = len(matriks)

    # Inisialisasi solusi dengan tebakan awal
    solusi = nilai_awal.copy()

    # Menampilkan header tabel
    print("Iterasi\t", end="")
    for i in range(n):
        print(f"x{i+1}\t\t", end="")
    print()

    # Menjalankan iterasi Gauss-Seidel
    for iterasi in range(max_iterasi):
        prev_solusi = solusi.copy()

        # Mengupdate solusi pada setiap variabel
        for i in range(n):
            sigma = 0.0
            for j in range(n):
                if j != i:
                    sigma += matriks[i][j] * solusi[j]

            solusi[i] = (vektor[i] - sigma) / matriks[i][i]

        # Menampilkan solusi pada setiap iterasi
        print(f"{iterasi+1}\t\t", end="")
        for i in range(n):
            print(f"{solusi[i]:.4f}\t", end="")
        print()

    # Menampilkan solusi akhir
    print("Solusi Akhir:")
    for i in range(n):
        print(f"x{i+1} = {solusi[i]:.4f}")

# Input matriks dari pengguna
n = int(input("Masukkan ukuran matriks (n): "))
matriks = []
print("Masukkan elemen-elemen matriks:")
for i in range(n):
    row = list(map(float, input().split()))
    matriks.append(row)

# Input vektor dari pengguna
print("Masukkan elemen-elemen vektor:")
vektor = list(map(float, input().split()))

# Input tebakan awal dari pengguna
print("Masukkan tebakan awal:")
nilai_awal = list(map(float, input().split()))

# Input maksimum iterasi dari pengguna
max_iterasi = int(input("Masukkan maksimum iterasi: "))

# Menjalankan metode Gauss-Seidel
gauss_seidel(matriks, vektor, nilai_awal, max_iterasi)
```
