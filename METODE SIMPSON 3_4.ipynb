```python
import random
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2

def simpson_3_4_rule(f, a, b, n):
    h = (b - a) / n
    print('nilai h : ', h)
    integral = f(a) + f(b)

    for i in range(1, n):
        x = a + i * h
        if i % 3 == 0:
            integral += 2 * f(x)
        else:
            integral += 3 * f(x)

    integral *= 3 * h / 8
    return integral

# Define the limits of integration
a = int(input('masukkan batas bawah :'))
b = int(input('masukkan batas atas :'))

# Number of segments
n = int(input('masukkan banyaknya segmen:'))

# Calculate the approximation using Simpson's 3/4 rule
approximation = simpson_3_4_rule(f, a, b, n)

print("Pendekatan integral:", approximation)

# Generate x values for the area plot
x = np.linspace(a, b, 100)
y = f(x)

# Plotting the area under the curve
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f(x)')
plt.fill_between(x, y, alpha=0.5, label='Area under Curve')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grafik f(x) dan Luasan Daerah')
plt.legend()
plt.show()
```
