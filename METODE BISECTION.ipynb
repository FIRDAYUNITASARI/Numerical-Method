```python
import math
import random
from prettytable import PrettyTable
import matplotlib.pyplot as plt

def f(x, y):
    return x - y + math.log(x)

def bisection_method(f, a, b, y, epsilon):
    if f(a, y) * f(b, y) >= 0:
        raise Exception("Tidak ada perpotongan sumbu x dalam interval ini.")

    table = PrettyTable()
    table.field_names = ["Iterasi", "a", "b", "c", "f(a)", "f(b)", "f(c)", "Galat"]

    iterasi = 0
    c = (a + b) / 2
    fc = f(c, y)
    galat = abs(b - a)

    table.add_row([iterasi, format(a, ".6f"), format(b, ".6f"), format(c, ".6f"),
                   format(f(a, y), ".6f"),format(f(b, y), ".6f"),
                   format(fc, ".6f"), format(galat, ".6f")])

    titik = [c]

    while galat > epsilon:
        if galat < epsilon:
            break

        if f(a, y) * fc < 0:
            b = c
        else:
            a = c

        iterasi += 1
        c = (a + b) / 2
        fc = f(c, y)
        galat = abs(b - a)

        table.add_row([iterasi, format(a, ".6f"), format(b, ".6f"), format(c, ".6f"),
                       format(f(a, y), ".6f"), format(f(b, y), ".6f"),
                       format(fc, ".6f"), format(galat, ".6f")])

        titik.append(c)

    print(table)

    return c, titik

# Input interval a dan b
a = float(input('Masukkan nilai a: '))
b = float(input('Masukkan nilai b: '))

# Input nilai y
y = random.uniform(2, 40)
print("Nilai y: ", y)

# Input galat (epsilon)
epsilon = float(input('Masukkan nilai epsilon: '))

# Mencari akar persamaan dengan metode biseksi
akar, titik = bisection_method(f, a, b, y, epsilon)

print("Akar persamaan: ", format(akar, ".6f"))

# Plot sumbu x adalah iterasi dan sumbu y adalah nilai c
plt.plot(titik, 'bo-')
plt.xlabel('Iterasi')
plt.ylabel('Nilai c')
plt.title('Plot Nilai c dan Iterasi')
plt.grid(True)
plt.show()
```
