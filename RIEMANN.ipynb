```python
import numpy as np
import matplotlib.pyplot as plt

def riemann_sum(a, b, equation, h):
    n = int((b - a) / h)  # Jumlah subinterval
    x = np.linspace(a, b, n+1)  # Titik-titik x pada interval [a, b]
    y = eval(equation)  # Menghitung nilai y pada setiap titik x

    total_area = 0  # Total luas

    print("x\t\tf(x)\t\tArea")

    for i in range(n):
        area = y[i] * h  # Menghitung luas subinterval
        total_area += area

        print("{:.2f}\t\t{:.2f}\t\t{:.2f}".format(x[i], y[i], area))

    print("Total area: {:.3f}".format(total_area))

    # Plot grafik fungsi
    plt.plot(x, y, 'r-', label='f(x) = (6*x)-(x**2)')

    # Plot persegi untuk representasi luas
    plt.bar(x, y, width=h, align='edge', alpha=0.5, label='Area')

    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Riemann Sum')
    plt.legend()
    plt.show()

# Contoh penggunaan
riemann_sum(0, 6, "(6*x)-(x**2)", 1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def riemann_sum(a, b, equation, h):
    n = int((b - a) / h)  # Jumlah subinterval
    x = a  # Titik awal
    total_area = 0  # Total luas
    areas = []  # List untuk menyimpan luas tiap subinterval

    print("x\t\tf(x)\t\tArea")

    for i in range(n):
        area = eval(equation) * h  # Menghitung luas subinterval
        total_area += area
        areas.append(total_area)  # Menyimpan luas tiap subinterval

        print("{:.2f}\t\t{:.2f}\t\t{:.2f}".format(x, eval(equation), area))

        x += h  # Pindah ke titik berikutnya

    print("Total area: {:.3f}".format(total_area))

    # Plot grafik luas daerah
    x_vals = np.linspace(a, b, n)  # Titik x pada interval [a, b]
    plt.plot(x_vals, areas, 'b-')

    plt.xlabel('x')
    plt.ylabel('Total Area')
    plt.title('Luas Daerah dengan Riemann Sum')
    plt.grid(True)
    plt.show()

# Contoh penggunaan
riemann_sum(0, 6, "(6*x)-(x**2)", 1)
```
